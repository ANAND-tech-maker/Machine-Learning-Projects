```python
X = df[['Flight_Distance_km', 'Departure_Hour', 'Arrival_Hour', 'Temperature_C',
       'Wind_Speed_kmph', 'Humidity_Percent', 'Previous_Delays',
       'Passenger_Count', 'Airline', 'Weather_Condition', 'Day_Type',
       'Airport', 'Flight_Type']]
```

```python
y = df["Delayed"]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train
```

```python
X_train_cat = X_train[["Airline","Weather_Condition","Day_Type"	,"Airport","Flight_Type"]]

X_train_num = X_train[["Flight_Distance_km",
                      "Departure_Hour",	"Arrival_Hour",
                      "Temperature_C",	
                      "Wind_Speed_kmph" ,
                      "Humidity_Percent",	
                      "Previous_Delays",
                      "Passenger_Count"]]
```

```python
from sklearn.preprocessing import LabelEncoder

X_train_cat_transformed = pd.DataFrame(index=X_train_cat.index)

for col in X_train_cat.columns:
    le = LabelEncoder()
    X_train_cat_transformed[col] = le.fit_transform(X_train_cat[col].astype(str))
```

```python
X_train_cat_transformed
```

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

scaler = StandardScaler()

X_train_num_transformed = pd.DataFrame(
    scaler.fit_transform(X_train_num),
    columns=X_train_num.columns,
    index=X_train_num.index
)
```

```python
X_train_num_transformed
```

```python
X_train_transformed = pd.concat([X_train_cat_transformed,X_train_num_transformed],axis=1)
```

```python
X_train_transformed
```

```python
X_test_cat = X_test.select_dtypes(include=["object"])
```

```python
X_test_cat
```

```python
X_test_cat = X_test.select_dtypes(include=['object'])
X_test_num = X_test.select_dtypes(include=['int64', 'float64'])
```

```python
X_test_num_transformed = pd.DataFrame(scaler.transform(X_test_num), 
                                   columns = scaler.get_feature_names_out(), 
                                   index = X_test_num.index)

X_test_num_transformed.head()
```

```python

X_test_cat_transformed = pd.DataFrame(index=X_test_cat.index)

for col in X_test_cat.columns:
    le = LabelEncoder()
    le.fit(X_train_cat[col].astype(str)) 
    X_test_cat_transformed[col] = le.transform(X_test_cat[col].astype(str))

X_test_cat_transformed.head()
```

```python
X_test_transformed = pd.concat([X_test_cat_transformed,X_test_num_transformed],axis=1)
```

```python
X_test_transformed
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000, random_state=42)

model.fit(X_train_transformed,y_train)
```

```python
y_pred = model.predict(X_test_transformed)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_pred,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

model = RandomForestClassifier()

model.fit(X_train_transformed,y_train)

y_pred = model.predict(X_test_transformed)


accuracy_score(y_pred,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

model = DecisionTreeClassifier()

model.fit(X_train_transformed,y_train)

y_pred = model.predict(X_test_transformed)


accuracy_score(y_pred,y_test)
```
